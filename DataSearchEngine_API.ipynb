```python
from fastapi import FastAPI
import nest_asyncio
from pyngrok import ngrok
import uvicorn
```

```python
from pydantic import BaseModel
```

```python
!ngrok config add-authtoken 2PdKi1m8fTESfuaP65fqbglxmmx_5qJmCoMgiy7q3akdJaxBm
```

```python
class Company_Request(BaseModel):
    company_name: str
    company_country: str
    company_URL: str | None = None
```

```python
class Company_Response(BaseModel):
    products_services: list | None = None
    keywords: list | None = None
    classification: list | None = None
    image_urls: list | None = None
```

```python
def generateResponse():
  response = Company_Response()
  response.products_services = ['Furniture', 'Home decor', 'Kitchen and Dining']
  response.keywords = ['furniture', 'storage', 'lighting']
  response.classification = ['5712 (Furniture Stores) – SIC', '442110 (Furniture Stores) – NAICS']
  response.image_urls = ['https://serpapi.com/searches/6457b0f296f5d7b407953423/images/6b515bb71cfd9544e871f1a84db4ce91f0326052b4ce8cfa5a6e7b75aa833d50.png',
                         'https://serpapi.com/searches/6457b0f296f5d7b407953423/images/6b515bb71cfd95446106ea0179b25564db13ecd5dcd73f523c665affcf84c56c.jpeg',
                         'https://serpapi.com/searches/6457b0f296f5d7b407953423/images/6b515bb71cfd95444d5f3a3d16fab7d0d9b0d1ce65f296ea6f1a4a9547b12e11.jpeg']
  return response
```

```python
app = FastAPI()
```

```python
@app.post("/company/")
async def create_item(item: Company_Request):
  company_response = generateResponse()
  return company_response
```

```python
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```
