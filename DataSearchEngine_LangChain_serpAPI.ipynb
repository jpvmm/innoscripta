```python
def googleSearch(search_prompt, SERPAPI_KEY, img_count):
  search = GoogleSearch({
    "q": search_prompt, 
    "engine": "google_images",
    "location": "Austin,Texas",
    "api_key": SERPAPI_KEY
  })
  response = search.get_dict()
  images_results_json = response['images_results']
  thumbnails_json = [x['thumbnail'] for x in images_results_json][:img_count]
  return thumbnails_json
```
